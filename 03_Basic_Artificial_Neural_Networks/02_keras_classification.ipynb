```python
X= df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.fit_transform(X_test)
```

## Creating the Model 

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# Bınary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

# Training the Model

### Example one: Choosing too many epochs and overfitting

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1)
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

## Example Two: Early Stopping

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

# Bınary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
model_loss.plot()
```

## Example Three: Adding in Dropout Layers

Dropout katmandaki bazı nöronların görmezden gelinerek eğitimin yapılmasıdır

```python
from keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5))  # 0.5 ve aşağısı önerilir. 

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
pred= (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```
